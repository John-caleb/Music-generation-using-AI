```python
octave_notes = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']
```

```python
def transpose_to_c(song, scale):
    notes = song.split(' ')
    new_scale = 'C'
    diff = 12 - octave_notes.index(scale)
    new_notes = []
    for i in notes:
        x = (octave_notes.index(i) + diff)%12
        new_notes.append(octave_notes[x])
    return new_notes
```
